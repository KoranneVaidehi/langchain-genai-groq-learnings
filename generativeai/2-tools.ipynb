## Tool execution loop

```python
from langchain.tools import tool
from langchain_core.messages import SystemMessage, HumanMessage

@tool("weather_tool")
def get_weather(location: str) -> str:
    """Get the weather for a location"""
    return f"The weather in {location} is sunny."

model_with_tools = model.bind_tools([get_weather])

messages = [
    SystemMessage(content="Always call weather_tool for weather questions."),
    HumanMessage(content="What is the weather in Bangalore?")
]

ai_msg = model_with_tools.invoke(messages)

print("Tool calls:", ai_msg.tool_calls)
```

```python
from langchain.tools import tool
from langchain_core.messages import SystemMessage, HumanMessage

@tool("weather_tool")
def get_weather(location: str) -> str:
    """Get the weather for a location"""
    return f"The weather in {location} is sunny."

model_with_tools = model.bind_tools([get_weather])

messages = [
    SystemMessage(content="Always use weather_tool for weather questions."),
    HumanMessage(content="What is the weather in Bangalore?")
]

ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)

# run tool
for tool_call in ai_msg.tool_calls:
    tool_result = get_weather.invoke(tool_call["args"])
    messages.append(tool_result)

# final response
final_response = model_with_tools.invoke(messages)
print(final_response.content)
```
