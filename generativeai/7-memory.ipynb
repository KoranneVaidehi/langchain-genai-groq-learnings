### Agents with memory

```python
from dotenv import load_dotenv
import os

from langchain_groq import ChatGroq

load_dotenv()

groq_key = os.getenv("GROQ_API_KEY")
print("GROQ_API_KEY Loaded:", "YES" if groq_key else "NO")

model = ChatGroq(
    model="llama-3.1-8b-instant",
    api_key=groq_key  # optional but recommended to avoid old key issue
)

response = model.invoke("Hello! Give me 1 line about Inception movie.")
print(response.content)
```
